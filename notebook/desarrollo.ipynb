```python
#Ejercicio 2

lista1 = [2, 5, 4, 7, 3]
lista2 = [1, 4, 6, 9, 10]
 
# Printing original lists
print ("Lista Original 1 : " + str(lista1))
print ("Lista Original 2 : " + str(lista2))
 
res_list = []
for i in range(0, len(lista1)):
    res_list.append(lista1[i] + lista2[i])
  
print ("Lista resultado es : " + str(res_list))
```

```python
# Comprensin de listas
 
# Printing original lists
print ("Lista Original 1 : " + str(lista1))
print ("Lista Original 2 : " + str(lista2))
 
# Using list comprehension to add two lists

res_list = [lista1[i]+lista2[i] for i in range(len(lista1))]
 
# Printing resulting list 

print( 'resulting list:' + str(res_list) )
```

```python
from operator import add
# Use map() + add():
# Initializing lists
# Printing original lists
# Using map() + add() to add two lists
res_list = list(map(add, lista1, lista2))
 
# Printing resulting list 
print ("Resulting list is : " + str(res_list))
```

```python
# Use zip() + sum(): esta funcin sirve no slo para mapear funciones sencillas, pueden ser complejas.
# Initializing lists
# Printing original lists
# Using zip() + sum() to add two lists
res_list = [sum(i) for i in zip(lista1, lista2)]
 
# Printing resulting list 
print ("Resulting list is : " + str(res_list))
```

```python
#Ejercicio 3

matrix1 = [[1,7,3],
 [4,5,2],
 [3,6,1]]
matrix2 = [[5,4,1],
 [1,2,3],
 [4,5,2]]

res = [[0 for x in range(3)] for y in range(3)]
 
# Explicit for loops
for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
        for k in range(len(matrix2)):
 
            # Resulting matrix
            res[i][j] += matrix1[i][k] * matrix2[k][j]
 
print(res)
```

```python


# Input two matrices
A=np.array(matrix1)
B=np.array(matrix2)
 
# This will return dot product
res = A.dot(B)
 
# Print resulting matrix
print(res)
```
